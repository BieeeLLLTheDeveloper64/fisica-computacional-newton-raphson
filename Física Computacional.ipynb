```python
import numpy as np
import matplotlib.pyplot as plt

# Função g1(u, v, Jx, Jy, t)
def g1(u, v, Jx, Jy, t):
    r1 = Jx * u + Jy * v
    return np.tanh(r1 / t)

# Função f1(Jx, Jy, t, m)
def f1(Jx, Jy, t, m):
    B00 = (g1(2.0, 2.0, Jx, Jy, t) / 16.0 + g1(-2.0, -2.0, Jx, Jy, t) / 16.0 +
           g1(2.0, -2.0, Jx, Jy, t) / 16.0 + g1(-2.0, 2.0, Jx, Jy, t) / 16.0 +
           g1(2.0, 0.0, Jx, Jy, t) / 8.0 + g1(-2.0, 0.0, Jx, Jy, t) / 8.0 +
           g1(0.0, 2.0, Jx, Jy, t) / 8.0 + g1(0.0, -2.0, Jx, Jy, t) / 8.0 +
           g1(0.0, 0.0, Jx, Jy, t) / 8.0)

    B01 = (g1(2.0, 2.0, Jx, Jy, t) / 8.0 - g1(-2.0, -2.0, Jx, Jy, t) / 8.0 -
           g1(2.0, -2.0, Jx, Jy, t) / 8.0 + g1(-2.0, 2.0, Jx, Jy, t) / 8.0 +
           g1(0.0, 2.0, Jx, Jy, t) / 4.0 - g1(0.0, -2.0, Jx, Jy, t) / 4.0)

    B02 = (-g1(2.0, 0.0, Jx, Jy, t) / 8.0 - g1(-2.0, 0.0, Jx, Jy, t) / 8.0 +
            g1(2.0, 2.0, Jx, Jy, t) / 16.0 + g1(-2.0, -2.0, Jx, Jy, t) / 16.0 +
            g1(2.0, -2.0, Jx, Jy, t) / 16.0 + g1(-2.0, 2.0, Jx, Jy, t) / 16.0 -
            g1(0.0, 0.0, Jx, Jy, t) / 8.0 + g1(0.0, 2.0, Jx, Jy, t) / 8.0 +
            g1(0.0, -2.0, Jx, Jy, t) / 8.0)

    B10 = (g1(2.0, 2.0, Jx, Jy, t) / 8.0 - g1(-2.0, -2.0, Jx, Jy, t) / 8.0 -
            g1(-2.0, 2.0, Jx, Jy, t) / 8.0 + g1(2.0, -2.0, Jx, Jy, t) / 8.0 +
            g1(2.0, 0.0, Jx, Jy, t) / 4.0 - g1(-2.0, 0.0, Jx, Jy, t) / 4.0)

    B11 = (g1(2.0, 2.0, Jx, Jy, t) / 4.0 + g1(-2.0, -2.0, Jx, Jy, t) / 4.0 -
            g1(2.0, -2.0, Jx, Jy, t) / 4.0 - g1(-2.0, 2.0, Jx, Jy, t) / 4.0)

    B12 = (-g1(2.0, 0.0, Jx, Jy, t) / 4.0 + g1(-2.0, 0.0, Jx, Jy, t) / 4.0 +
            g1(2.0, 2.0, Jx, Jy, t) / 8.0 - g1(-2.0, -2.0, Jx, Jy, t) / 8.0 -
            g1(-2.0, 2.0, Jx, Jy, t) / 8.0 + g1(2.0, -2.0, Jx, Jy, t) / 8.0)

    B20 = (-g1(0.0, 2.0, Jx, Jy, t) / 8.0 - g1(0.0, -2.0, Jx, Jy, t) / 8.0 -
            g1(0.0, 0.0, Jx, Jy, t) / 8.0 + g1(2.0, 2.0, Jx, Jy, t) / 16.0 +
            g1(-2.0, -2.0, Jx, Jy, t) / 16.0 + g1(-2.0, 2.0, Jx, Jy, t) / 16.0 +
            g1(2.0, -2.0, Jx, Jy, t) / 16.0 + g1(2.0, 0.0, Jx, Jy, t) / 8.0 +
            g1(-2.0, 0.0, Jx, Jy, t) / 8.0)

    B21 = (-g1(0.0, 2.0, Jx, Jy, t) / 4.0 + g1(0.0, -2.0, Jx, Jy, t) / 4.0 +
            g1(2.0, 2.0, Jx, Jy, t) / 8.0 - g1(-2.0, -2.0, Jx, Jy, t) / 8.0 -
            g1(2.0, -2.0, Jx, Jy, t) / 8.0 + g1(-2.0, 2.0, Jx, Jy, t) / 8.0)

    B22 = (g1(0.0, 0.0, Jx, Jy, t) / 8.0 - g1(0.0, 2.0, Jx, Jy, t) / 8.0 -
            g1(0.0, -2.0, Jx, Jy, t) / 8.0 - g1(2.0, 0.0, Jx, Jy, t) / 8.0 -
            g1(-2.0, 0.0, Jx, Jy, t) / 8.0 + g1(2.0, 2.0, Jx, Jy, t) / 16.0 +
            g1(-2.0, -2.0, Jx, Jy, t) / 16.0 + g1(2.0, -2.0, Jx, Jy, t) / 16.0 +
            g1(-2.0, 2.0, Jx, Jy, t) / 16.0)

    return m - B00 - m * (B01 + B10) - m**2 * (B02 - B11 - B20) - m**3 * (B12 + B21) - m**4 * B22

# Programa principal
def main():
    Jx = 1.0
    Jy = 0.2
    m = 1.0
    del_val = 1.0e-4
    epson = 1.0e-5
    t_values = []
    m_values = []

    with open('jy1.dat', 'w') as file:
        for t in np.arange(0.001, 4.0, 0.001):
            achou = False
            for n in range(20):
                ff1 = f1(Jx, Jy, t, m)

                if abs(ff1) <= epson:
                    t_values.append(t)
                    m_values.append(m)
                    file.write(f"{t} {m}\n")
                    achou = True
                    break

                # Cálculo da derivada numérica de f1 em relação a m
                df1m = (f1(Jx, Jy, t, m + del_val) - f1(Jx, Jy, t, m - del_val)) / (2.0 * del_val)

                # Atualiza m usando o método de Newton-Raphson
                m = m - (ff1 / df1m)

            if not achou:
                continue

    # Gerando o gráfico
    plt.figure(figsize=(8, 6))
    plt.plot(t_values, m_values, 'b-', label='m vs t')
    plt.xlabel('t')
    plt.ylabel('m')
    plt.title('Gráfico de m em função de t')
    plt.grid(True)
    plt.legend()
    plt.savefig('m_vs_t.png')
    plt.show()

if __name__ == "__main__":
    main()
```
